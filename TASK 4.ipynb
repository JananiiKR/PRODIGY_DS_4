```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example sentiment data
sentiments = ['Positive', 'Negative', 'Neutral', 'Positive', 'Negative', 'Positive', 'Neutral', 'Negative', 'Positive']
sentiment_counts = {'Positive': sentiments.count('Positive'), 'Negative': sentiments.count('Negative'), 'Neutral': sentiments.count('Neutral')}

# Bar chart for sentiment distribution
plt.figure(figsize=(6,4))
sns.barplot(x=list(sentiment_counts.keys()), y=list(sentiment_counts.values()), palette="Blues_d")
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example time-series data
data = {
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Sentiment_Score': [0.5, 0.2, 0.7, -0.3, 0.8, -0.1, 0.6, -0.2, 0.4, 0.9]
}

df = pd.DataFrame(data)

# Line chart to show sentiment over time
plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['Sentiment_Score'], marker='o', color='b')
plt.title('Sentiment Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sentiment Score')
plt.grid(True)
plt.show()
```
