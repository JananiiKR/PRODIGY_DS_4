```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for age distribution
ages = [22, 25, 30, 32, 35, 35, 40, 42, 50, 55, 60, 65, 70, 75, 80, 85, 90]

# Create histogram for age distribution
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=8, edgecolor='black', color='skyblue')
plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
